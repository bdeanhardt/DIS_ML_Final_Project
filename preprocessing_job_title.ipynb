Exploration

```python
import pandas as pd
pd.set_option('display.max_rows', None)

data = pd.read_csv('DataScience_salaries_2024.csv')

filter_threshold = 0.01

job_data = data['job_title'].value_counts().reset_index()
job_data.columns = ['job_title', 'frequency']
job_data = job_data[(job_data['frequency'] / job_data.shape[0]) >= filter_threshold]
job_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
job_title_vec = vectorizer.fit_transform(job_data['job_title']).toarray()

from sklearn.cluster import KMeans

n_clusters = 8

kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(job_title_vec)
predict = kmeans.predict(job_title_vec)
for i in range(10):
  for j in range(len(job_title_vec)):
    if predict[j] == i:
      print(predict[j], ': ', job_data['job_title'][j])
```

Preprocessing Function

```python
def process_job_title(job_titles):
  for i, title in enumerate(job_titles):
     title = title.replace('ML', 'Machine Learning').replace('BI', 'Business Intelligence').replace('AI', 'Artificial Intelligence').replace('Science', 'Scientist')
    
  job_cat = ['Data', 'Machine Learning', 'Business Intelligence', 'Artificial Intelligence', 'Research']
  job_type = ['Engineer', 'Analyst', 'Scientist']

  def check_job_cat_or_type(jobs, title):
    for job in jobs:
       if job in title:
          return job
    return None

  for i, title in enumerate(job_titles):
     title = title.replace('ML', 'Machine Learning').replace('BI', 'Business Intelligence').replace('AI', 'Artificial Intelligence')

  for i, title in enumerate(job_titles):
    check1 = check_job_cat_or_type(job_cat, title)
    check2 = check_job_cat_or_type(job_type, title)

    if check1 is not None:
        title = check1 + ' '
    else:
        title = 'Data '
    
    if check2 is not None:
        title += check2
    else:
        title += 'Scientist'
    
    job_titles[i] = title

    return job_titles
```
